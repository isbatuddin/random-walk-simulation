Visualize all walks

````python
# NumPy is imported, seed is set
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(123)

# Initialize all_walks list
all_walks = []

# Simulate random walk 5 times
for i in range(5):  # Loop 5 times for 5 random walks
    random_walk = [0]  # Start each walk at 0
    
    for x in range(100):  # 100 steps in each random walk
        step = random_walk[-1]
        dice = np.random.randint(1, 7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1, 7)

        random_walk.append(step)
    
    # Append the completed random_walk to all_walks
    all_walks.append(random_walk)

# Convert all_walks to a NumPy array
np_aw = np.array(all_walks)

# Plot the random walks using plt.plot()
plt.plot(np_aw)
plt.show()

# Transpose np_aw to have each row representing the position after 1 throw for the five random walks
np_aw_t = np.transpose(np_aw)

# Plot the transposed array
plt.plot(np_aw_t)
plt.show()
plt.clf()
````

Implement clumsiness

````python
# numpy and matplotlib imported, seed set
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(123)

# Clear the plot so it doesn't get cluttered if you run this many times
plt.clf()

# Simulate random walk 20 times
all_walks = []
for i in range(20):  # Run the simulation 20 times
    random_walk = [0]  # Start each walk at step 0
    for x in range(100):  # 100 steps per random walk
        step = random_walk[-1]
        dice = np.random.randint(1, 7)  # Roll the dice

        # Update step based on dice roll
        if dice <= 2:
            step = max(0, step - 1)  # Move one step down
        elif dice <= 5:
            step = step + 1  # Move one step up
        else:
            step = step + np.random.randint(1, 7)  # Roll again if dice is 6

        # Implement clumsiness: if random float is <= 0.005, set step to 0
        if np.random.rand() <= 0.005:
            step = 0

        # Append the updated step to the random walk
        random_walk.append(step)
    
    # Append each random_walk to the all_walks list
    all_walks.append(random_walk)

# Convert all_walks to a NumPy array and transpose it
np_aw_t = np.transpose(np.array(all_walks))

# Plot the transposed random walks
plt.plot(np_aw_t)
plt.show()
````
